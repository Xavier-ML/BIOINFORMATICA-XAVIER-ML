```python
import matplotlib.pyplot as plt
from ramachandraw.utils import plot

# Define the PDB file paths
pdb_filepaths = ["template.pdb","serine_ba.B99990001.pdb","serine_ba.B99990002.pdb",
                 "serine_ba.B99990003.pdb","serine_ba.B99990004.pdb","serine_ba.B99990005.pdb"
]

# Loop through each PDB file and generate a Ramachandran plot
for filepath in pdb_filepaths:
  plot(filepath)
  plt.title(f'Ramachandran Plot for {filepath}')
  plt.show()
  # Clear the plot for the next iteration (optional)
  plt.clf()

print("Ramachandran plots generated for all PDB files.")
```

```python
import matplotlib.pyplot as plt
from ramachandraw.utils import plot

# Define colors for each PDB file
colors = ['b', 'g', 'r', 'c', 'm', 'o']  # Adjust the color list for your needs

# Define the PDB file paths
pdb_filepaths = ["template.pdb","serine_ba.B99990001.pdb","serine_ba.B99990002.pdb",
                 "serine_ba.B99990003.pdb","serine_ba.B99990004.pdb","serine_ba.B99990005.pdb"
]

# Create an empty figure using plt.figure()
fig = plt.figure()

# Loop through PDB files and plot on the same axes with different colors
for i, filepath in enumerate(pdb_filepaths):
    # Create a new subplot within the figure
    ax = fig.add_subplot(1, len(pdb_filepaths), i + 1)
    # Plot the data on the current axes
    plot(filepath, ax=ax, color=colors[i], alpha=0.75)

# Set labels and title using plt methods
ax.set_xlabel('Phi')
ax.set_ylabel('Psi')
plt.title('Ramachandran Plot for Multiple Proteins')

# Save and show the plot (optional)
plt.savefig('ramachandran_plot.png', dpi=100)
plt.show()
```
